```python
 import matplotlib.pyplot as plt
import seaborn as sns

# style
sns.set_style("whitegrid")

# columns
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# grid setup
n_cols = 3
n_rows = (len(cols) + n_cols - 1) // n_cols

# bigger figure to fit long labels
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 12))
axes = axes.flatten()

# plotting
for i, col in enumerate(cols):
    sns.countplot(data=df, x=col, ax=axes[i],hue="Churn")
    
    # title fix
    axes[i].set_title(col, pad=12)
    
    # 👇 KEY FIX: keep labels straight (no rotation)
    axes[i].tick_params(axis='x', rotation=0)
    
    # optional: center align labels
    for label in axes[i].get_xticklabels():
        label.set_horizontalalignment('center')

# remove extra axes
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# spacing adjustments
plt.subplots_adjust(hspace=0.4, wspace=0.3)
plt.tight_layout()

plt.show()
```

```python
 #Customers without security, tech support, backup, or protection—especially fiber optic users—have the highest churn, while customers with these services or no internet service show significantly lower churn, and phone/streaming features have minimal impact.
```
